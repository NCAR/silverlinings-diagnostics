# Plot the Output

## Imports

```python
import matplotlib.pyplot as plt
import xarray as xr
from datetime import datetime
```

## Read in the Data

```python
aws_ds = xr.open_dataset('data/global_average_temperature_aws_year1.nc')
cheyenne_ds = xr.open_dataset('data/global_average_temperature.nc').sel(time='2035')
```

```python
cheyenne_ds.time
```

## Plot the Output

```python
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111)

aws_ds.TREFHT.plot(color='black', lw=3, label='AWS run')
cheyenne_ds.TREFHT.plot(hue='case')

plt.xlabel('Year')
```
