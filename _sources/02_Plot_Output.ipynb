# Plot the Output

## Imports

```python
import matplotlib.pyplot as plt
import xarray as xr
```

## Read in the Data

```python
ds = xr.open_dataset('data/global_average_temperature.nc')
```

## Plot the Output

```python
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111)

ds.TREFHT.plot(hue='case')

plt.xlabel('Year')

plt.ylim(285, 290)
```
